```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("amazon.csv")
```

```python
#Histogram of price column

plt.figure(figsize=(10,6))
sns.histplot(df['rating_count'], bins=20, kde=True)
plt.title("Rating Count")
plt.show()
```

```python
#boxplot of the ratings
plt.figure(figsize=(10,6))
sns.boxplot(x=df['rating'])
plt.title('Rating Distribution', color="red", fontsize=20)
plt.show
```

```python
#scattering of price vs units sold
plt.figure(figsize=(10,6))
sns.scatterplot(x='discounted_price', y='actual_price', data=df)
plt.title('price vs unit sold')
plt.show
```

```python
#barplot of categories vs avg units sold
plt.figure(figsize=(10,6))
sns.barplot(x='category', y='discounted_price', data=df)
plt.title("Avg unit sold by category")
plt.xticks(rotation=90)
plt.show
```
